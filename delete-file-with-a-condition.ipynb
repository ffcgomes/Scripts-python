```python
import os
import promptlib
import pdfplumber
import pandas as pd
#Abre o diretório onde estão os laudos a serem lidos
prompter = promptlib.Files()
dir = prompter.dir()
print(dir)
```

```python
#Mostra os laudos contidos no diretório
files=os.listdir(dir)
files
```

```python
#Conta a quantidade laudos 
n_laudos=len(files)
n_laudos
```
